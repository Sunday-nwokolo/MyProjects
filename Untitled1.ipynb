```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a synthetic reference image
reference = np.zeros((100, 100))
reference[40:60, 40:60] = 1

# Step 2: Create a target image by shifting the reference image 5 pixels to the right
target = np.zeros_like(reference)
target[40:60, 45:65] = 1

# Visualize reference and initial target
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(reference, cmap='gray')
plt.title('Reference Image')

plt.subplot(1, 3, 2)
plt.imshow(target, cmap='gray')
plt.title('Target Image (Shifted)')

# Step 3: Define a fitness function
def fitness(shift):
    shifted_target = np.roll(target, shift=(int(shift[0]), int(shift[1])), axis=(0, 1))
    error = np.sum((reference - shifted_target) ** 2)
    return error

# Step 4: Use an optimization algorithm
from scipy.optimize import minimize

result = minimize(fitness, [0, 0], method='Powell')

# Display the best shifted target
best_shifted_target = np.roll(target, shift=(int(result.x[0]), int(result.x[1])), axis=(0, 1))
plt.subplot(1, 3, 3)
plt.imshow(best_shifted_target, cmap='gray')
plt.title('Best Shifted Target')

plt.tight_layout()
plt.show()

print(f"Estimated translation: {result.x}")
```
