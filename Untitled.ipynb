```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color

# Load the reference image
reference = io.imread('C:/Users/snn23kfl/project/captured_images/ref.jpg', as_gray=True)

# Convert image to grayscale if it is RGB
if len(reference.shape) > 2:
    reference = color.rgb2gray(reference)

# Artificially shift the reference image to create the target image
true_shift = [20, -15]  # y-shift of 20 and x-shift of -15 for demonstration purposes
target = np.roll(reference, shift=true_shift, axis=(0, 1))

# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(reference, cmap='gray')
plt.title('Reference Image')
plt.subplot(1, 2, 2)
plt.imshow(target, cmap='gray')
plt.title('Artificially Shifted Target Image')
plt.show()

# Define a fitness function
def fitness(shift):
    shifted_ref = np.roll(reference, shift=(-int(shift[0]), -int(shift[1])), axis=(0, 1))
    error = np.sum((shifted_ref - target) ** 2)
    return error

# Use an optimization algorithm
from scipy.optimize import minimize

result = minimize(fitness, [0, 0], method='Powell')

print(f"True translation applied: {true_shift}")
print(f"Estimated translation: {result.x}")
```
